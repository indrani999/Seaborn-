# Seaborn
##### Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for creating attractive and informative statistical graphics. It comes with several built-in themes and color palettes to make it easy to create visually appealing plots. Seaborn is particularly useful for visualizing complex datasets with multiple variables.

##### Seaborn supports various types of charts, and here are some of the most commonly used ones:

1. Relational Plots:

scatterplot: Displays individual data points in a

*   Scatterplot: Displays individual data points in a two-dimensional space.
*   Lineplot: Draws lines to connect individual data points.

2. Categorical Plots:

*   Barplot: Shows the central tendency of a numerical variable with height of bars.
*   Countplot: Displays the count of observations in each category as bars.
*   Boxplot: Represents the distribution of categorical data.
*   Violinplot: Combines aspects of a box plot and a kernel density estimate.

3. Distribution Plots:

*   Histplot: Shows the distribution of a single variable.
*   Kdeplot: Displays the Kernel Density Estimate of a univariate distribution.
*   Ecdfplot: Plots empirical cumulative distribution functions.

4. Matrix Plots:

*   Heatmap: Displays a 2D dataset as a color-encoded matrix.
*   Clustermap: Clusters and orders rows and columns of a matrix dataset.
5. Regression Plots:

*   Regplot: Plots a scatterplot with a linear regression model fit.
*   Lmplot: Combines regplot with FacetGrid to allow for easy plotting on multiple subplots.

6. Multi-plot Grids:

*   FacetGrid: Allows for multiple plots on one figure.

##### Seaborn provides a high-level interface to create these plots with minimal code, and it is especially useful for exploring relationships in complex datasets. It also integrates well with Pandas DataFrames, making it easy to work with tabular data.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
spotify_filepath = ("/content/spotify.csv")

# Read the file into a variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
# Print the first 5 rows of the data
spotify_data.head()
```

```python
# Print the last five rows of the data
spotify_data.tail()
```

```python
# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
# Path of the file to read
flight_filepath =("/content/flight_delays.csv")

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
# Print the data
flight_data
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

```python
# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)
```

```python
# Path of the file to read
iris_filepath = ("/content/iris (1).csv")

# Read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Print the first 5 rows of the data
iris_data.head()
```

```python
# Histogram
sns.histplot(iris_data['Petal Length (cm)'])
```

```python
# KDE plot
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
# Histograms for each species
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

# Add title
plt.title("Histogram of Petal Lengths, by Species")
```

```python
# KDE plots for each species
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)

# Add title
plt.title("Distribution of Petal Lengths, by Species")
```
